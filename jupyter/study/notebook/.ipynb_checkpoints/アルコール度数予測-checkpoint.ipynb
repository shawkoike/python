### データ準備

UCI が公開しているワインのデータセットを使用する。
今回はワインデータセットの「アルコール度数」を予測する回帰タスクを例として実装する。

```python
import pandas as pd
import matplotlib.pyplot as plot
import numpy as np
%matplotlib inline
```

```python
#  ワインのデータセットの読み込み
# 赤ワインと白ワインの二つに分かれているので統合する
df_red = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv", sep=";")
df_white = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv", sep=";")

# 赤ワインは０、白ワインは１でラベルをつけておく
df_red["class"] = 0
df_white["class"] = 1
# データの統合
df_all = pd.concat([df_red, df_white]).reset_index(drop=True)
# alcohol カラムを目的変数とみなす
# 説明変数
X_df = df_all.drop("alcohol", axis=1)
# 目的変数
y_df = df_all.alcohol
```

### パイプラインの実装法

機械学習ではモデルを作成する前に前処理のステップを行う場合がある。     
例えば、以下の例ではPCAによる次元削減と正規化という２つの前処理のステップがある。

① PCAにより次元削減     
② 特徴量を正規化     
③ SVMによりモデル作成       

モデル作成時に必要な全てのステップを簡潔にできるようにする仕組みがパイプラインである。     
パイプラインでは基本的に sklearn の書き方に従ったモジュールであればどのようなライブラリでも利用可能であるうえ、自作モジュールも利用できる。

SVM を利用した実装例が以下である。

```python
# パイプライン
from sklearn.pipeline import Pipeline
# バリデーションに必要なライブラリ群
from sklearn.model_selection import train_test_split
# 前処理に必要なライブラリ群
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
# サポートベクターマシン
from sklearn.svm import SVR
```

```python
# trainデータと testデータに分割
(X_train, X_test, y_train, y_test) = train_test_split(X_df, y_df, test_size=0.3, random_state=0)

# パイプラインを活用
# 実行したいステップをまとめる
steps_svm = [
    ('pca', PCA(n_components=3, random_state=0)), # PCA で３次元まで次元削減
    ('sc', StandardScaler()), # StandardScalerによりデータを正規化
    ('svm', SVR()), # SVR（サポートベクターマシンの回帰モデル）によりモデル作成
]

# パイプラインの宣言
pipe_svm = Pipeline(steps_svm)

# 全てのステップを一度に実行
pipe_svm.fit(X_train, y_train)

# X_test, y_test でデータの予測を行う
pipe_svm.score(X_test, y_test)
```

### グリッドサーチの実装

グリッドサーチは、ハイパーパラメータの調整及びクロスバリデーションをまとめて行うことができる。     
グリッドサーチを実行することで自動的にベストスコアとなるハイパーパラメータを選択してくれる。

またグリッドサーチとパイプラインを同時に利用することで、リークを起こすことなくパラメータのチューニング及びモデルの評価を行うことができる。

```python
#クロスバリデーションに必要なライブラリ群
from sklearn.model_selection import StratifiedKFold,KFold

#パイプライン、グリッドサーチ
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
```

```python
#パイプラインを活用
#実行したいステップをまとめる
steps_svm = [
    ('pca', PCA(n_components=3,random_state=0)), #PCAにより３次元まで次元を削減
    ('sc', StandardScaler()), #StandardScalerによりデータを正規化
    ('svm', SVR()), #SVR(サポートベクターマシンの回帰モデル)によりモデル作成
]

#パイプラインの宣言
pipe_svm = Pipeline(steps_svm)
#チューニングしたいハイパーパラメータを記述
#「モジュール名__ハイパーパラメータ名」というルールで記述
params_svm = {
    'pca__n_components' : [1,2,3], #PCAの次元数を1,2,3の中から最適なものを選ぶ
    'svm__gamma' : [0.01,0.1,1], #svmのガンマを調整    
    'svm__C' : [5.0,2.0,1.0],  #svmのCパラメータを調整    
    'svm__kernel' : ['rbf','linear'], #svmのカーネルを調整
}
#クロスバリデーションの宣言
#今回は5分割に交差検証し、グリッドサーチを行う
folds = KFold(n_splits=5, shuffle=True,random_state=40)
#グリッドサーチの宣言
#scoringはパラメータの評価指標であり、今回はR2を利用
svm = GridSearchCV(pipe_svm, params_svm, scoring="r2", cv=folds)
#クロスバリデーション、パラメータチューニング、パイプラインを一括で実行する
svm.fit(X_train,y_train) 
```

```python
GridSearchCV(cv=KFold(n_splits=5, random_state=40, shuffle=True),
             error_score='raise-deprecating',
             estimator=Pipeline(memory=None,
                                steps=[('pca',
                                        PCA(copy=True, iterated_power='auto',
                                            n_components=3, random_state=0,
                                            svd_solver='auto', tol=0.0,
                                            whiten=False)),
                                       ('sc',
                                        StandardScaler(copy=True,
                                                       with_mean=True,
                                                       with_std=True)),
                                       ('svm',
                                        SVR(C=1.0, cache_size=200, coef0=0.0,
                                            degree=3, epsilon=0.1,
                                            gamma='auto_deprecated',
                                            kernel='rbf', max_iter=-1,
                                            shrinking=True, tol=0.001,
                                            verbose=False))],
                                verbose=False),
             iid='warn', n_jobs=None,
             param_grid={'pca__n_components': [1, 2, 3],
                         'svm__C': [5.0, 2.0, 1.0],
                         'svm__gamma': [0.01, 0.1, 1],
                         'svm__kernel': ['rbf', 'linear']},
             pre_dispatch='2*n_jobs', refit=True, return_train_score=False,
             scoring='r2', verbose=0)
```

### スタッキングの実装法

スタッキングとは様々な学習機で予測した結果を用いて再度モデルを作成することである。

なるべく相関が低い学習機同士を組み合わせることで汎化性能も上がることが知られており、様々な学習機を組み合わせてみることが大切。

スタッキングの基本は、クロスバリデーションの時と同様に学習用データをn分割することである。
クロスバリデーションの時と違う点は、学習用で得られたモデルを利用して残りのデータをpredictする。

全ての学習用データをpredictした結果を利用し、再度モデルを作成することがスタッキングの基本。 クロスバリデーションのようにデータを分割して予測することで、データリークを防ぐことが可能。

### スタッキング前準備

スタッキングを実行する前に準備としてグリッドサーチとパイプラインを利用してモデルを定義する。

今回はなるべく多くの sklearnモデルを読み込み、様々な学習機の値でスタッキングする。

```python
#sklearnで利用できるモデルのライブラリ例
## サポートベクターマシン
from sklearn.svm import SVR
## バギング、ブースティング系
from sklearn.ensemble import RandomForestRegressor,AdaBoostRegressor,ExtraTreesRegressor,GradientBoostingRegressor
from xgboost.sklearn import XGBRegressor
from lightgbm.sklearn import LGBMRegressor
## 線形モデル
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso,Ridge,ElasticNet
## ニューラルネットワーク
from sklearn.neural_network import MLPRegressor
## k近傍法
from sklearn.neighbors import KNeighborsRegressor
```

```python
def get_pipe_line(steps):
    return Pipeline(steps)

def get_grid_search(model,params,scoring="r2"):
    #今回は3分割に交差検証し、グリッドサーチを行う
    folds = KFold(n_splits=3, shuffle=True,random_state=0)
    return GridSearchCV(model, params, n_jobs=-1, verbose=-1, scoring=scoring, cv=folds)
```

```python
#################################### サポートベクターマシン ####################################
steps_svm = [
    #('pca', PCA(n_components=100,random_state=0)),　 ※先程の例では精度があまり良くなかった為、今回はPCAは行いません。
    ('sc', StandardScaler()),
    ('svm', SVR(kernel='rbf')),
]
params_svm = {
    'svm__gamma' : [0.01,0.1,1],
    'svm__C' : [5.0,2.0,1.0],
    'svm__kernel' : ['rbf'],
}
#パイプラインの定義
pipe_svm = get_pipe_line(steps_svm)
#グリッドサーチの定義
svm = get_grid_search(pipe_svm, params_svm)


#################################### xgb ####################################
steps_xgb = [
    ('xgb', XGBRegressor()),
]
params_xgb = {
    'xgb__random_state' : [0],
    'xgb__n_estimators' : [1000],
    'xgb__max_depth' : [2],
    'xgb__learning_rate' : [0.09],
}
#パイプラインの定義
pipe_xgb = get_pipe_line(steps_xgb)
#グリッドサーチの定義
xgb = get_grid_search(pipe_xgb, params_xgb)


#################################### Extra Tree ####################################
steps_et = [
    ('et', ExtraTreesRegressor(random_state=0)),
]
params_et = {
    'et__n_jobs': [-1],
    'et__n_estimators' : [1000],
    'et__max_features': [0.5],
    'et__max_depth': [8,None],
    'et__min_samples_leaf': [2],
    'et__verbose': [0],
}
#パイプラインの定義
pipe_et = get_pipe_line(steps_et)
#グリッドサーチの定義
et = get_grid_search(pipe_et, params_et)


#################################### Lightgbm ####################################
steps_lgbm = [
    ('lgbm', LGBMRegressor()),
]
params_lgbm = {
    'lgbm__random_state' : [0],
    'lgbm__n_estimators' : [10000],
}
#パイプラインの定義
pipe_lgbm = get_pipe_line(steps_lgbm)
#グリッドサーチの定義
lgbm = get_grid_search(pipe_lgbm, params_lgbm)


#################################### Ada Boost ####################################
steps_ada = [
    ('ada', AdaBoostRegressor(random_state=0)),
]
params_ada = {
    'ada__n_estimators' : [1000],
    'ada__learning_rate' : [0.01],
    'ada__base_estimator' : [None],
}
#パイプラインの定義
pipe_ada = get_pipe_line(steps_ada)
#グリッドサーチの定義
ada = get_grid_search(pipe_ada, params_ada)


#################################### Gradient Boost ####################################
steps_gb = [
    ('gb', GradientBoostingRegressor(random_state=0)),
]
params_gb = {
    'gb__n_estimators' : [1000],
    'gb__learning_rate' : [0.01],
}
#パイプラインの定義
pipe_gb = get_pipe_line(steps_gb)
#グリッドサーチの定義
gb = get_grid_search(pipe_gb, params_gb)


#################################### Random Forest ####################################
steps_forest = [
    ('forest', RandomForestRegressor(n_jobs=-1)),
]
params_forest = {
    'forest__random_state' : [0],
    'forest__n_estimators' : [1000],
    'forest__max_depth' : [None],
    'forest__n_jobs' : [-1]
}
#パイプラインの定義
pipe_forest = get_pipe_line(steps_forest)
#グリッドサーチの定義
forest = get_grid_search(pipe_forest, params_forest)


#################################### LinearRegression ####################################
steps_linear = [
    ('linear', LinearRegression()),
]
params_linear = {
    #パラメータがない場合は定義しなくても大丈夫です
}
#パイプラインの定義
pipe_linear = get_pipe_line(steps_linear)
#グリッドサーチの定義
linear = get_grid_search(pipe_linear, params_linear)


#################################### Lasso ####################################
steps_lasso = [
    ('lasso',Lasso()),
]
params_lasso = {
    'lasso__alpha' : [0.0009],
    'lasso__max_iter' : [10000],
}
#パイプラインの定義
pipe_lasso = get_pipe_line(steps_lasso)
#グリッドサーチの定義
lasso = get_grid_search(pipe_lasso, params_lasso)


#################################### Ridge ####################################
steps_ridge = [
    ('ridge', Ridge()),
]
params_ridge = {
    'ridge__max_iter' : [1000],
}
#パイプラインの定義
pipe_ridge = get_pipe_line(steps_ridge)
#グリッドサーチの定義
ridge = get_grid_search(pipe_ridge, params_ridge)


#################################### Elastic Net ####################################
steps_erastic = [
    ('sc', StandardScaler()),
    ('erastic', ElasticNet()),
]
params_erastic = {
    'erastic__random_state' : [0],
    'erastic__max_iter' : [10000],
    'erastic__alpha' : [0.0009],
}
#パイプラインの定義
pipe_erastic = get_pipe_line(steps_erastic)
#グリッドサーチの定義
erastic = get_grid_search(pipe_erastic, params_erastic)


#################################### ニューラルネットワーク ####################################
steps_mlp = [
    ('sc', StandardScaler()),
    ('mlp', MLPRegressor(random_state=0)),
]
params_mlp = {
    'mlp__random_state' : [0],
    'mlp__learning_rate_init' : [0.0015,0.001],
    'mlp__hidden_layer_sizes' : [(2,2),(1,1),(1),(2),(3)],
    'mlp__max_iter' : [10000],
}
#パイプラインの定義
pipe_mlp = get_pipe_line(steps_mlp)
#グリッドサーチの定義
mlp = get_grid_search(pipe_mlp, params_mlp)


#################################### k近傍法 ####################################
steps_knn = [
    ('sc', StandardScaler()),
    ('knn', KNeighborsRegressor()),
]
params_knn = {
    'knn__n_jobs' : [-1],
    'knn__n_neighbors' : [1,2,3,4,5],
}
#パイプラインの定義
pipe_knn = get_pipe_line(steps_knn)
#グリッドサーチの定義
knn = get_grid_search(pipe_knn, params_knn)
```

```python
all_model={
    "xgb" : xgb,
    "et" : et,   
    "mlp" : mlp,
    "lgbm" : lgbm,
    "knn" : knn,
    "linear" : linear,
    "lasso" : lasso,
    "svm" : svm,
    "forest" : forest,
    "ridge" : ridge,
    "ada" : ada,
    "gb" : gb,
    "erastic" : erastic,
}
```

```python
#予測結果を格納しておく箱を用意する
#とりあえず最初はターゲットの値のみを格納する
stack_train = pd.DataFrame(y_train)
stack_test = pd.DataFrame(y_test)
#クロスバリデーション分のモデルの平均を格納しておくための箱
stack_test_mean = pd.DataFrame(y_test)

#全てのモデルに対して最初は予測結果にnullを定義しておく、
for key,model in all_model.items():
    #train用
    stack_train[key]=np.nan
    #test用
    stack_test_mean[key]=np.nan    
    #クロスバリデーションの回数だけループを回す。今回はfold=3回分
    fold = 3
    for i in range(0,fold):
        col_nm = str(key)+str(i)
        #test用
        stack_test[col_nm]=np.nan
```

### スタッキングの実行

まずは通常のモデル作成方法でスコアを出力

```python
#all_modelのモデルに対して通常のモデル作成を行ってみる
#Target変数はy_trainおよびy_test（アルコール度数）

#通常の予測
normal_model_res = {}
for key,model in all_model.items():
        model.fit(X_train,y_train)
        #スタッキングの結果も配列に格納しておく
        normal_model_res[key] = model.score(X_test,y_test)

#通常のハイスコアモデルを記録しておく
max_normal_model = max(normal_model_res.items(), key=lambda x: x[1])
```

次にスタッキングのやり方で予測を行う

```python
#X_trainを5分割し、５回分のスタッキングを各々のモデルで行う
kf = KFold(n_splits=5,random_state=30,shuffle=True)

#単体モデルでの予測実行
single_model_res = {}
#X_trainの学習データから、predictデータとX_testデータの予測を行う
for key,model in all_model.items():
    col_nm_list=[]
    i=0
    #X_trainを学習用データとpredictデータに５分割する
    for X_train_stack,X_test_stack in kf.split(X_train):
        #4分割分でモデルを作成
        model.fit(X_train.iloc[X_train_stack],y_train.iloc[X_train_stack])  
        #1分割分のデータを予測
        stack_train[key].iloc[X_test_stack] = model.predict(X_train.iloc[X_test_stack])
        
        #５個分のX_testの予測データをstack_testに格納
        col_nm = str(key)+str(i)
        col_nm_list.append(col_nm)
        stack_test[col_nm]=model.predict(X_test)
        #モデル単体の予測スコアを配列にまとめておく
        single_model_res[key] = model.score(X_test,y_test)
        
        i += 1
    #５分割分の予測結果の平均をまとめる
    stack_test_mean[key] = stack_test[col_nm_list].mean(axis='columns')
    
#スタッキングの単体モデルでのハイスコアを記録しておく
max_single_model = max(single_model_res.items(), key=lambda x: x[1])
```

今回は特徴量自体も少なく、またデータ自体もそれ程多くないためか、モデル同士の相関もある程度高い結果となっています。例えば、linearとmlpモデル同士や、linearとerasticモデル同士は特にモデル間の今回は特徴量自体も少なく、またデータ自体もそれ程多くないためか、モデル同士の相関もある程度高い結果となっています。例えば、linearとmlpモデル同士や、linearとerasticモデル同士は特にモデル間の相関性が高く、色も白色に近くなっています。
一方lassoモデルやridgeモデルは、このモデル間での相関性は高いですが、その他のモデルとの相関性は低く赤色に近い色で表現されています。

一点注意点として、スタッキングで利用するモデルの基準として、相関性を利用した指標はありません。あるモデルをスタッキングに含めるべきかどうかの判断は、実際にそのモデルを除いた場合と含めた場合の精度で評価することが現状一般的なようです。

モデル同士の相関性は、どれほどスタッキングモデルに多様性があるかを確認する為に利用しましょう。モデル同士の相関性が低いと、スタッキングにより作成したモデルは多様性が大きくなり、より大きな表現力を持つ（汎化性能が上がる）と期待できます。相関性が高く、色も白色に近くなっています。
一方lassoモデルやridgeモデルは、このモデル間での相関性は高いですが、その他のモデルとの相関性は低く赤色に近い色で表現されています。

以上で単体モデルでの予測を計算することができました。では、それらの結果であるstack_train、stack_testを利用して再度ターゲットの予測を行ってみましょう

```python
#all_modelを再利用して予測します
#特徴量は、先程予測した結果であるstack_trainおよびstack_test_mean
#Target変数は先ほどと変わらずy_trainおよびy_test（アルコール度数）

#ターゲット変数であるalcoholは削除
stack_train.drop("alcohol",axis=1,inplace=True)
stack_test_mean.drop("alcohol",axis=1,inplace=True)

#スタッキングでの予測
stack_model_res = {}
for key,model in all_model.items():
        model.fit(stack_train,y_train)
        #スタッキングの結果も配列に格納しておく
        stack_model_res[key] = model.score(stack_test_mean,y_test)

#スタッキングのハイスコアを記録しておく
max_stack_model = max(stack_model_res.items(), key=lambda x: x[1])
```

単体の場合と同様に、スタッキング結果のベストモデルを確認してみましょう。下の結果は決定係数：R2により各手法のスコアを出力した結果です。R2スコアは通常0~1の値を取り、1に近いほど高精度です。

スタッキングの結果では、２層目のモデルにsvmを利用して予測した結果が最も高精度で約0.921となっています。また、通常手法の場合はスコアが約0.911であり、２層スタッキングの結果が最も高精度であることがわかります。

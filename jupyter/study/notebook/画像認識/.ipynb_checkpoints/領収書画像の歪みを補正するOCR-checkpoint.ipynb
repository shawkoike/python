OpenCV を使用して領収書の画像の歪みを補正し、Google が提供する光学文字認識エンジン「Tesseract OCR」を使用して OCR を行い、活字の文書の画像を文字コードの列に変換してテキスト文字を抽出する。

まず OpenCV を使用して領収書の画像の歪みを補正。一般的に徴収書などはユーザーの携帯のカメラなどで取られることが多く、画像が少し傾いていたりする。
そのような場合に、OCR処理を行うと精度が低くなるため、事前に画像処理を行って OCR で読み取りやすい形式に変換する。

OpenCV を使用して、画像の歪みを補正した後に Tesseract OCR をを使用して OCR処理をかける
Tesseract PCR は Google が管理しているオープンソースの OCRライブラリである。

今回は、実際に歪みのある領収書の画像を歪みの補正前と補正後でどの程度 OCR 精度に差があるのか確認する。

### 歪み補正前の徴収書を OCR 実行

````python
import os
import cv2
import pytesseract
import numpy as np

from matplotlib import pyplot as plt
````

````python
# Tesseract パスを指定
pytesseract.pytesseract.tesseact_cmd = r'/usr/local/bin/tesseract'

# 画像の読み込み
filepath = os.path.join(os.getcwd(), "img/sample.png")
image = cv2.imread(filepath)

# 画像の表示
from IPython.display import Image, display_png
display_png(Image(filepath))
````

````python
# OCR処理
# 言語は日本語を選択
# ページセグメンテーションは６を選択（単一のテキストブロックとみなして、横書きの日本語のときに高い精度が期待できる）
raw_text = pytesseract.image_to_string(image, lang="jpn", config="--psm 6")
print(raw_text)
````

### OpenCV を使用して画像の歪みを補正

画像の歪み補正の手順

1. 画像の読み込み
2. 画像をグレースケール化して、色反転を行う
3. 画像を二値化
4. ピクセル座標から回転角を取得
5. 歪みを補正
6. 画像の保存

````python
# 画像をグレースケール化して、色反転を行う
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray_image = cv2.bitwise_not(gray_image)

plt.imshow(gray_image)
plt.show()
````

````python
# 画像の二値化
thresh_image = cv2.threshold(gray_image, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]

# 全てのピクセルで0より大きい値の座標を取得し、「cv2.minAreaRect」によって回転角を取得
coords = np.column_stack(np.where(thresh_image > 0))
angle = cv2.minAreaRect(coords)[-1]

#「cv2.minAreaRect」は、[-90, 0)の領域で値を返すため、回転角を調整する
if angle < -45:
	angle = -(90 + angle)
else:
	angle = -angle

# 画像の歪みを補正
(h, w) = image.shape[:2]
center = (w // 2, h // 2)
matrix = cv2.getRotationMatrix2D(center, angle, 1.0) # 回転変換行列の算出
rotated_image = cv2.warpAffine(image, matrix, (w, h), flags=cv2.INTER_CUBIC, borderMode=cv2.BORDER_REPLICATE) # 画像のアフィン変換

plt.imshow(rotated_image)
plt.show()
````

````python
# 画像の保存
cv2.imwrite("./img/rotated.png", rotated_image)
````

### 歪み補正後の領収書を OCRしてみる

````python
# 画像の読み込み
filepath = os.path.join(os.getcwd(), "./img/rotated.png")

image = cv2.imread(filepath)

# OCR処理
# 言語は日本語を選択
# ページセグメンテーションは6を選択（単一のテキストブロックとみなして、横書きの日本語のときに高い精度が期待できる）
raw_text = pytesseract.image_to_string(image, lang="jpn", config="--psm 6")
print(raw_text)
````

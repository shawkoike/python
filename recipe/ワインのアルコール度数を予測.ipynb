```python
import pandas as pd
import matplotlib.pyplot as plot
import numpy as np
%matplotlib inline
```

```python
# ワインのデータセットの読み込み
df_red = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv", sep=";")
df_white = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv", sep=";")

# 赤ワインは0、白ワインは1でラベルをつけておく
df_red["class"] = 0
df_white["class"] = 1

# データの統合
df_all = pd.concat([df_red, df_white]).reset_index(drop=True)
# alcohol が目的変数
X_df = df_all.drop("alcohol", axis=1)
y_df = df_all.alcohol
```
